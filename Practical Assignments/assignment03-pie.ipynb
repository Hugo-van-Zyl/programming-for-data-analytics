```python
# Import pandas that I can pull in CSV data to a dataframe
import pandas as pd

# Import matplotlib library that I can use for creating charts, in particular pie chart
import matplotlib as plt

# Import seaborn library. I will use their standard selection of color palettes to color my pie chart
import seaborn as sns

# Get CSV data into a pandas dataframe. CSV file is in this root folder so no need to specify filepath
df_peopledata = pd.read_csv("people-1000.csv")

# display first few rows of data to make sure it pulled into dataframe correctly
# display(df_peopledata.head())

# Get / isolate the email addresses only from the larger dataframe
# Split the domain name from the people specific email address using the string split() method
# use .str[1] to keep everything after the delimeter @
df_domains = df_peopledata['Email'].str.split('@').str[1]

# check output if correct
# display(df_domains)

# Now I need to get the total number for each unique domain in the datafrane. I will use the values_count() method for this
df_domain_counts = df_domains.value_counts()

# check output if correct
# display(df_domain_counts) 

# Choose a color palette from seaborn for using in my pie chart
pie_colors = sns.color_palette('hls')

# Now create the pie chart using this data
piechart = df_domain_counts.plot(kind='pie', autopct='%1.2f%%', colors= pie_colors, figsize=(6,6), shadow = True)
# set shadow = True to make it look nicer. 

# set a title for the pie chart
piechart.set_title('Percentage of domain names across 1000 people sample')
# no y label here as this is pie chart
piechart.set_ylabel('')

```
