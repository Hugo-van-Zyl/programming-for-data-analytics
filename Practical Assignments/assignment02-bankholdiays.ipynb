```python
# Import this module that allows me to send HTTP requests
import requests 

# set variable to address of bank holidays data as provided in link from our lecturer
url_bankholidays = "https://www.gov.uk/bank-holidays.json" 

# Send a get request to my specified address to return the object
response = requests.get(url_bankholidays) 

data_bankholidays = response.json() 

# print the data to see what format it returns. 
print(data_bankholidays)
# this data is shown as a dictionary file. Therefore, I can select specific data by

#print(data['northern-ireland']) 



```
