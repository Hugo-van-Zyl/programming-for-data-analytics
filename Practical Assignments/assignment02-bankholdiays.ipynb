# Task Summary: 
Write a program called assignment02-bankholdiays.py

The program should print out the dates of the bank holidays that happen in northern Ireland.

Modify the program to print the bank holidays that are unique to northern Ireland (i.e. do not happen elsewhere in the UK) you can choose if you want to use the name or the date of the holiday to decide if it is unique.

```python
# Import this module that allows me to send HTTP requests
import requests 

# Import pandas that I can work with dataframes. Note to lecturer - Make sure you have pandas installed using !pip install pandas on your compiler
# import pandas as pd
# Note I tried to do this firts with pandas but failed, so I moved on to using lists

# set variable to address of bank holidays data as provided in link from our lecturer
url_bankholidays = "https://www.gov.uk/bank-holidays.json" 

# Send a get request to my specified address to return the object
# Furthermore, I know this is a JSON object so I will add.json() to return the JSON object
data_bankholidays = requests.get(url_bankholidays).json() #I dont need another variable to set data_bankholidays = response.json() like in the weekly lab as I did it in one step above 

# print the data to see what format it returns. 
# print(data_bankholidays)
# print(data_bankholidays.keys())
# this data is shown as a dictionary file. Therefore, I can select specific data by checking the keys()

# Now I select the bank holidays for nothern ireland only as the task required
print("The bank holidays that occur in Nothern Ireland are : ")
print(data_bankholidays['northern-ireland']) # this outputs the dictionary object for bank holidays in northern ireland.

# Now I have to print out names of bank holidays unique to nothern ireland
# First I will print all bank holidays for each region from the possible keys I checked above and I will use the holiday name to check for duplicates
bankholidays_englandwales = data_bankholidays['england-and-wales']['events']
#print(bankholidays_ew) # Test the output to see if correct
bankholidays_scotland = data_bankholidays['scotland']['events']
# print(bankholidays_s) # Test the output to see if correct
bankholidays_northernireland = data_bankholidays['northern-ireland']['events']
# print(bankholidays_ni) # Test the output to see if correct

# create a blank list where I will store all the bank holiday names for ew
titles_englandwales = []
# use for loop to iterate through each title in the dictionary and append this to the blank list
for event in bankholidays_englandwales:
    titles_englandwales.append(event['title']) # add title to the list

print("Bank holiday names for england and wales are :", '\n',set(titles_englandwales)) # use set() to remove duplicates from multiple years data

# create a blank list where I will store all the bank holiday names for ew
titles_scotland = []
# use for loop to iterate through each title in the dictionary and append this to the blank list
for event in bankholidays_scotland:
    titles_scotland.append(event['title']) # add title to the list

print("Bank holiday names for scotland are :", '\n',set(titles_scotland)) # use set() to remove duplicates from multiple years data

# create a blank list where I will store all the bank holiday names for ew
titles_northernireland = []
# use for loop to iterate through each title in the dictionary and append this to the blank list
for event in bankholidays_northernireland:
    titles_northernireland.append(event['title']) # add title to the list

print("Bank holiday names for northern ireland are :", '\n',set(titles_northernireland)) # use set() to remove duplicates from multiple years data

# use the Python set method to remove all duplicates in each list as they have multiple years of data and then to compare the lists with northern ireland and remove any further duplicates with scotland and england&wales
unique_northernireland = set(titles_northernireland)-set(titles_scotland)-set(titles_englandwales)
print("\n", "Bank holidays unique to Northern Ireland are: ",unique_northernireland)
```
